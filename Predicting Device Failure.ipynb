### Undersampling

```python
from sklearn.model_selection import train_test_split as tts
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import cross_val_score as cvs
```

```python
X = devicefail.drop('failure', axis = 1)
Y = devicefail['failure']
```

```python
X.shape
```

```python
Y.shape
```

```python
rus = RandomUnderSampler(random_state = 0)
```

```python
Xres, Yres= rus.fit_sample(X,Y)
```

```python
Xres.shape
```

```python
Yres.shape
```

```python
Xtrain, Xtest, Ytrain, Ytest = tts(Xres,Yres, test_size = .2, random_state = 0, stratify = Yres)
```

```python
sh = [Xtrain, Xtest, Ytrain, Ytest]

for i in range(len(s)):
    print(s[i].shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
l = LogisticRegression()
d = DecisionTreeClassifier()
r = RandomForestClassifier()
g = GradientBoostingClassifier()
k = KNeighborsClassifier (n_neighbors = 3)
```

```python
lr = l.fit(Xtrain, Ytrain)
dc = d.fit(Xtrain, Ytrain)
rc = r.fit(Xtrain, Ytrain)
gc = g.fit(Xtrain, Ytrain)
kc = k.fit(Xtrain, Ytrain)
```

```python
cl = [lr,dc,rc,gc,kc]
alg = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'Gradient Booster','Kneighbors']

for i in range(len(alg)):
    print(alg[i])
    print('Precision of RUS, ', cvs(cl[i],Xres,Yres,cv=5, scoring='precision'))
    print('Precision is, ', cvs(cl[i],Xres,Yres,cv=5, scoring='precision').mean())
    print('Recall of RUS ,', cvs(cl[i],Xres,Yres,cv=5, scoring='recall'))
    print('Recall is, ', cvs(cl[i],Xres,Yres,cv=5, scoring='recall').mean())
```

```python
lrpred = lr.predict(Xtest)
dcpred = dc.predict(Xtest)
rcpred = rc.predict(Xtest)
gcpred = gc.predict(Xtest)
kcpred = kc.predict(Xtest)
```

```python
col = [lrpred,dcpred,rcpred,gcpred,kcpred]
alg = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'Gradient Booster','Kneighbors']

for i in range(len(alg)):
    print(alg[i])
    print('Predict', precision_score(Ytest, col[i]))
    print('Recall' , recall_score(Ytest, col[i]))
    print('F1 score', f1_score(Ytest, col[i]))
    print(classification_report(Ytest, col[i]))
    print(confusion_matrix(Ytest, col[i]))
```
